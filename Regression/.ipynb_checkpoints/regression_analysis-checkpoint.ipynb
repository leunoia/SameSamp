```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

# Load Data

```python
pf = pd.read_csv('./sampled_music.csv')
df = pf.drop(pf.columns[[0,1,2,3,6,10,11,12,13]], axis=1)

df.head(10)
```

# Determine if any values in the data frame is missing

```python
df.isna().sum() # df.isnull().sum also return the number of missing values, you can use either of them]
```

# Data Preparation

```python
df = pd.get_dummies(df, columns=['genre', 'instrument'])
df['explicit'] = df['explicit'].astype(int)
df.drop_duplicates()
```

```python
X = df.drop('annualRev', axis=1)
y = df['annualRev']
```

## Splitting Training Sets & Training Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
rf_regressor = RandomForestRegressor()
```

```python
rf_regressor.fit(X_train, y_train)
```

```python
y_pred = rf_regressor.predict(X_test)
```

```python
score = rf_regressor.score(X_test, y_test)
score
```
